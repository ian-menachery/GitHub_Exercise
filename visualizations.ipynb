```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


titanic_data = sns.load_dataset('titanic')
print(titanic_data.head())
```

```python

sns.barplot(data=titanic_data, x="class", y="survived", hue="sex")
plt.title("Survival Rate by Class and Sex")
plt.ylabel("Survival Rate")
plt.xlabel("Class")
plt.legend(title="Sex")
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic_data, x="class", y="fare", hue="survived")
plt.title("Fare Distribution by Class and Survival")
plt.ylabel("Fare")
plt.xlabel("Class")
plt.legend(title="Survived", loc="upper right")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic_data, x="age", y="fare", hue="survived", palette="coolwarm")
plt.title("Age vs. Fare with Survival")
plt.ylabel("Fare")
plt.xlabel("Age")
plt.legend(title="Survived")
plt.show()
```
